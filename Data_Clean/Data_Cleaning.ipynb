```python
# import required libraries
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# read dataset
dataset = pd.read_csv(r'C:\Users\Administrator\Downloads\NFL\NFL.csv')
pd.set_option('display.max_columns', None)
dataset.head()
```

```python
# find shape of dataset
dataset.shape
```

```python
# display all columns' names
dataset.columns
```

```python
# find missing values if any in the dataset
print(dataset.iloc[:, 0:10].isnull().sum())
```

```python
# find percentage of missing values
total_values = len(dataset) * len(dataset.columns)
total_null_values = dataset.isnull().sum().sum()

# percentage of values
missing_percentage = (total_null_values/total_values)*100
print("percentage of missing values is", round(missing_percentage, 2), '%')
```

```python
# display missing values of first 10 columns
dataset.iloc[:, :10].isnull().sum()
```

```python
# fill timesecs column with mean of rest of the values in column
mean_value = dataset['TimeSecs'].mean()

# fill the missing values
dataset['TimeSecs'].fillna(mean_value, inplace = True)

dataset.iloc[:, :10].isnull().sum()
```

```python
# # find data types in time column
# dataset['time'] = dataset['time'].str.replace('14:53', '14.53')
# dataset['time'] = [str(time).replace(':', '.') for time in dataset['time']]
# dataset['time'] = dataset['time'].astype(float)
```

```python
# convert string data types in time column into float
dataset['time'] = dataset['time'].apply(lambda x: str(x).replace(':', '.') if isinstance(x, (str, float)) else x)
```

```python
# fill time column with median values
median_value = dataset['time'].median()
median_value
```

```python
# fill null values in time column with median vlaues
dataset['time'].fillna(median_value, inplace = True)
dataset.iloc[:, :10].isnull().sum()
```

```python
# # dispertion of column PlayTimeDiff for linearity or non linearity
# import plotly.express as px
# fig = px.scatter(dataset['PlayTimeDiff'], x = dataset['PlayTimeDiff'].index, y = 'PlayTimeDiff')
# fig.update_layout(title = 'Distribution of PlayTimeDiff', xaxis_title = 'Index', yaxis_title = 'PlayTimeDiff')
# fig.show()
```

```python
# interpolate missing values for column playTimeDiff
dataset['PlayTimeDiff'] = pd.to_numeric(dataset['PlayTimeDiff'], errors = 'coerce')

dataset['PlayTimeDiff'] = dataset['PlayTimeDiff'].interpolate(method = 'linear')

dataset.iloc[:, :10].isnull().sum()
```

```python
# forward fill values in column down
dataset['down'] = dataset['down'].fillna(method = 'ffill', inplace = False)

dataset.iloc[:, :10].isnull().sum()
```

```python
# we have one value is null in down so lets take mood of whole column and then replace in null cell
mode = dataset['down'].mode()

dataset['down'] = dataset['down'].fillna(mode, inplace = False)

dataset.iloc[:, :10].isnull().sum()
```

```python
# now fill values in sideoffield with backward fill method
dataset['SideofField'] = dataset['SideofField'].fillna(method = 'bfill', inplace = False)

dataset.iloc[:, :10].isnull().sum()
```

```python
# now show columns ranging from 11 to 20
dataset.iloc[:, 10:30].isnull().sum()
```

```python
# add median value of column yrd1n to the missing values in the column
median = dataset['yrdln'].median()

dataset['yrdln'] = dataset['yrdln'].fillna(median, inplace = False)

dataset.iloc[:, 10:30].isnull().sum()
```

```python
# lets now predict missing values in TwoPointConvo
missing_vals = dataset[dataset['TwoPointConv'].isnull()]

not_missing_vals = dataset[dataset['TwoPointConv'].notnull()]

not_missing_vals['TwoPointConv'].head()
```
