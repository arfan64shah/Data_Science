```python
# check the name column
titles = training_dataset.Name.str.extract(' ([A-Za-z]+)\. ', expand = False)
titles.unique()
```

```python
# i will not drop name column as i wanna use it to create a rank column and then analyze the data based on their ranks
training_dataset['Titles'] = titles
```

```python
# now drop Name column
training_dataset = training_dataset.drop('Name', axis = 1)
```

```python
# now handle sex with one hot encoding
training_dataset = pd.get_dummies(training_dataset, columns = ['Sex'])
training_dataset.head()
```

```python
# convert newly created Sex_female and Sex_male into integers
training_dataset.Sex_female = training_dataset['Sex_female'].astype(int)
training_dataset.Sex_male = training_dataset['Sex_male'].astype(int)
```
