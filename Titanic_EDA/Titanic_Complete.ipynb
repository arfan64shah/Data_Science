```python
# now create a temporary dataset from the original dataset
age_dataset = training_dataset[['Has_Cabin', 'Titles_Mr', 'Titles_Miss', 'SibSp', 'Pclass', 'Titles_Master', 'Age']]
```

```python
# display age dataset
age_dataset
```

```python
# now split age dataset into training and testing
age_testing = age_dataset[age_dataset['Age'].isnull()]

age_training = age_dataset[~age_dataset['Age'].isnull()]
age_training
```

```python
# now split training into dependent and independent
x_train = age_training[['Has_Cabin', 'Titles_Mr', 'Titles_Miss', 'SibSp', 'Pclass', 'Titles_Master']]
y_train = age_training['Age']
```

```python
y_train
```

```python
# split testing into dependent and independent
x_test = age_testing[['Has_Cabin', 'Titles_Mr', 'Titles_Miss', 'SibSp', 'Pclass', 'Titles_Master']]
```

```python
x_test
```

```python
# now use KNN algorithm to predict the remaining age values
lm = LinearRegression()
lm.fit(x_train, y_train)
```

```python
age_missing = lm.predict(x_test)
age_missing
```
