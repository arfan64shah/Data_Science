```python
# import required libraries
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
# load and read dataset
dataset = pd.read_csv('stock_data.csv')
dataset.head()
```

```python
dataset.columns
```

```python
# find all the dates
dataset['Date']
```

```python
# find numbe of instances
len(dataset)
```

```python
# show columns
dataset.columns
```

```python
# convert date column into date format
from datetime import datetime
dataset['Date'] = dataset['Date'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y'))
dataset.head()
```

```python
# now set date column as index
dataset.set_index('Date', inplace = True)
dataset.head()
```

```python
# remove index label from first row
dataset.index.name = None
dataset.head()
```

```python
# show scatter plot of Open with respect to time
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x = dataset.index, y = dataset['Open'], mode = 'markers', marker = dict(color = 'purple')))
fig1.update_layout(title = 'Distribution of Open', xaxis_title = 'Time', yaxis_title = 'Change in Open W.R.T Time', height = 500)
fig1.show()
```

```python
fig2 = go.Figure()
fig2.add_trace(go.Line(x = dataset.index, y = dataset['Open'], marker = dict(color = 'green')))
fig2.update_layout(title = 'Distribution of Open', xaxis_title = 'Time', yaxis_title = 'Open', height = 500)
fig2.show()
```

```python
# show distribution of Close
fig3 = go.Figure()
fig3.add_trace(go.Line(x = dataset.index, y = dataset['Close'], marker = dict(color = 'black')))
fig3.update_layout(title = 'Distribution of Close', xaxis_title = 'Time', yaxis_title = 'Close', height = 500)
fig3.show()
```

```python
# distribution of high
fig4 = go.Figure()
fig4.add_trace(go.Line(x = dataset.index, y = dataset['High'], marker = dict(color = 'red')))
fig4.update_layout(title = 'Distribution of High', xaxis_title = 'Time', yaxis_title = 'High', height = 500)
fig4.show()
```

```python
# check distribution of Low
fig5 = go.Figure()
fig5.add_trace(go.Line(x = dataset.index, y = dataset['Low'], marker = dict(color = 'gray')))
fig5.update_layout(title = 'Distribution of Low', xaxis_title = 'Time', yaxis_title = 'Low', height = 500)
fig5.show()
```

```python
# now show distribution of Volume
fig6 = go.Figure()
fig6.add_trace(go.Line(x = dataset.index, y = dataset['Volume'], marker = dict(color = 'darkblue')))
fig6.update_layout(title = 'Distribution of Volume', xaxis_title = 'Time', yaxis_title = 'Volume', height = 500)
fig6.show()
```
