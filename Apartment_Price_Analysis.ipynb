```python
# import requried libraries
import pandas as pd
import numpy as np

# read the dataset
data = pd.read_csv('house_price.csv')
data.head()
```

```python
# we do not need columns such as address, and latitudes and logitudes, so lets remove them
dataset = data.drop(['Unnamed: 0', 'address', 'date', 'latitude, longitude'], axis = 1)

# now display the new dataset
dataset.head()
```

```python
# find the length of dataset
len(dataset)
```

```python
# find other about dataset
dataset.info()
```

```python
# find out the relation between number of rooms and price
import matplotlib.pyplot as plt
plt.scatter(dataset['number_of_rooms'], dataset['price'], color = 'blue')
plt.xlabel('Number of Rooms')
plt.ylabel('Price')
plt.title('No of Rooms VS Price')
plt.show()
```

```python
# find out the relation between number of rooms and price
import matplotlib.pyplot as plt
plt.bar(dataset['number_of_rooms'], dataset['price'], color = 'blue')
plt.xlabel('Number of Rooms')
plt.ylabel('Price')
plt.title('No of Rooms VS Price')
plt.show()
```
