```python
# import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import plotly.graph_objects as go
```

```python
# Generate data points along the standard normal distribution (mean=0, std=1)
x = np.linspace(-3, 3, 1000)
y = norm.pdf(x, 0, 1)  # PDF for the standard normal distribution

# Plot the normal distribution
plt.plot(x, y, label='Standard Normal Distribution')

# Highlight specific Z-scores if needed
z_scores = [0, 1, -1]
colors = ['r', 'g', 'b']
for z, color in zip(z_scores, colors):
    plt.axvline(z, color=color, linestyle='--', label=f'Z = {z}')

# Add labels and legend
plt.title('Normal Distribution with Z-scores')
plt.xlabel('Z-score')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()

# Show the plot
plt.show()
```

```python
# create a function to calcuate z-score
def z_score(x, x_mu, std):
    z_score = (x - x_mu)/std
    return z_score
```

```python
print("z score of the respective person is", z_score(5.19, 3, 2))
```

```python
# z score for brushtail
print("z score is ", z_score(95.4, 92.6, 3.6))
```

```python
print("new z score is ", z_score(85.8, 92.6, 3.6))
```

```python
# Ann is student who scored 1300 in SAT and mean and std are 1100 and 200 respectively
ann_z_score = z_score(1300, 1100, 200)

print("Ann's Z-Score is ", ann_z_score)
```

```python
# find proportion of people below her in overall SAT
students_below = norm.cdf(ann_z_score)
print("Proportion of Students below Ann are ", students_below)
```

```python
print("Percentage of students below Ann are", students_below*100, "%")
```

```python
# remaioning above students
students_above = norm.sf(ann_z_score)

print("Percentage of students above Ann is", students_above*100, "%")
```
